```python
x=df['x']
```

```python
y=df['y']
```

```python
x.head()
```

```python
y.head()
```

```python
m=0
c=0
learning_rate=0.0001
n=int(len(x))
```

```python
n
```

```python
df.shape
```

```python
for i in range(1000): #itaration as much as better
  pred = m*x+c
  der_m = (-2/n)*sum(x*(y-pred))
  der_c = (-2/n)*sum(y-pred)

  m = m - learning_rate *der_m
  c = c - learning_rate *der_c

print(m,c)# m and c in global optima
```

```python
y = m*38.816818+c
y
```
