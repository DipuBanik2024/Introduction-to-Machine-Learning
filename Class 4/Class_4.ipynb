# One Hot Encoding using Loop

```python
df7.head()
```

```python
cols=['Area']
for col in cols:
  dummy1=pd.get_dummies(df7[col],prefix='Area',drop_first=True,dtype=int)
  df8=pd.concat([df7,dummy1],axis=1).drop(col,axis=1)
```

#Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df9.head()
```

```python
city=df9.Area.unique()
```

```python
city
```

```python
od=OrdinalEncoder(categories=[city])
```

```python
encoded_values=od.fit_transform(df9[['Area']])
encoded_values
```

```python
encoded_values=pd.DataFrame(encoded_values,columns=['Area'])
encoded_values.head()
```

```python
df9=pd.concat([df9.drop("Area",axis=1),encoded_values],axis=1)
df9.head()
```
