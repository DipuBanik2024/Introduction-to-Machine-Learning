# Lable Encoder


```python
df.columns
```

```python
from pandas.core.dtypes.common import is_numeric_dtype
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

for col in df.columns:
  if is_numeric_dtype(df[col]):
    continue
  else:
    df[col]=le.fit_transform(df[col])
```

```python
df.head()
```

# Scaling

```python
from sklearn.preprocessing import MinMaxScaler
mm=MinMaxScaler()
```

```python
df['Height']=mm.fit_transform(df[['Height']])
```

```python
df.head()
```

# Separete x,y

```python
x=df.drop('Weight',axis=1)
```

```python
x.head()
```

```python
y=df[['Weight']]
```

```python
y.head()
```

# Split Train and Test

```python
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LinearRegression
```

```python
x_train,x_test,y_train,y_test =tts(x,y,test_size=0.3,random_state=43)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
x_train.head()
```

```python
y_train.head()
```

# Apply Linear Regression

```python
ln=LinearRegression()
```

```python
ln.fit(x_train,y_train)
```

```python
pred=ln.predict(x_test)
pred
```

```python
ln.predict([[0,0.356]])
```

#Accuracy,MSE,MAE


```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
accuracy_score=r2_score(y_test,pred)
accuracy_score
```

```python
mae=mean_absolute_error(y_test,pred)
mae
```

```python
mse=mean_squared_error(y_test,pred)
mse
```

# Apply KNN Regression

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
ne=KNeighborsRegressor(n_neighbors=5)   #n_neighbors = 5 by defult
```

```python
ne.fit(x_train,y_train)
```

```python
kntest=ne.predict(x_test)
kntest
```

```python
ne.predict([[0,0.356]])
```

# Evaluate Model


```python
r2_score(y_test,kntest)
```

```python
mse=mean_squared_error(y_test,kntest)
mse
```

```python
mae=mean_absolute_error(y_test,kntest)
mae
```
