```python
import pandas as pd
from sklearn.datasets import make_classification
#https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_classification.html
#https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_regression.html
```

```python
x,y=make_classification(n_samples=10000,n_features=30)#for regrassion type data make_Regrassion
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
xtrain,xtest,ytrain,ytest=tts(x,y,test_size=0.3,random_state=42) #test_size e chaile number o diya daya jay{test_size=3000}
```

```python
xtest.shape  #sklearn the j dataset produce koreche eita datafram akare nai ,eita nmpy array from e aseche tai head() use kora jabe na
```

```python
ytest.shape
```

```python
xtest
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
train_score=[]
test_score=[]
```

```python
max_depth = [n for n in range(1,30)]
max_depth
```

```python
for depth in max_depth:
  clf=DecisionTreeClassifier(max_depth=depth)
  clf=clf.fit(xtrain,ytrain)

  #training accuracy
  train_acc = clf.score(xtest,ytest)
  train_score.append(train_acc)

  #test accuracy
  test_acc = clf.score(xtest,ytest)
  test_score.append(test_acc)
  print('Max depth = %d, training_accuracy = %f, test_accuracy = %f'%(depth,train_acc,test_acc))
```

```python
from matplotlib import pyplot as plt
plt.plot(max_depth, train_score, '-o', label = 'Train')
plt.plot(max_depth, test_score, '-o', label= 'Test')
plt.legend()
```
