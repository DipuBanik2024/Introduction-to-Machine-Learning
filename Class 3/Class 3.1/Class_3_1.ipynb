```python
import gradio as gr
```

```python
def rent_recom(x):
    x = float(x)
    rent = reg1.predict([[x]])
    return float(rent[0])
```

```python
rent_recom(3111)
```

```python

interface=gr.Interface(
    fn=rent_recom,
    inputs=gr.Number(),
    outputs=gr.Number()
)
interface.launch(share=False)
```
