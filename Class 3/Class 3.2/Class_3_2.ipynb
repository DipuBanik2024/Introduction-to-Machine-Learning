# Robust Scaler
# Robust Scaler are robust to outliers. It is used to scale the feature to median and quantiles Scaling using  median and quantiles consists of subtracting the median to all the observations, and then dividing by the  interquartile difference. The interquartile difference is the difference between the 75th and 25th quantile:
# IQR = 75th quantile - 25th quantile
# RobustScaler= (Xi â€“ X.Median)/IQR

```python
dataset=[8,9,2,10,3,5,7,12,15]
```

```python
Q1=np.percentile(dataset,25)
```

# Normalization

```python
from sklearn.preprocessing import MinMaxScaler
mm=MinMaxScaler(feature_range=(2,5))
```

```python
df1.head()
```

```python
df1['Marketing Spend']=mm.fit_transform(df1[['Marketing Spend']])
df1.head()
```

```python
df1['Administration']=mm.fit_transform(df1[['Administration']])
df1.head()
```

```python
df1['Transport']=mm.fit_transform(df1[['Transport']])
df1.head()
```

```python
df1['Profit']=mm.fit_transform(df1[['Profit']])
df1.head()
```
